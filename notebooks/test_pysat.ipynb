```python
import pysat.formula as pf
import numpy

n = 100
number_of_measurements = 150
vpool = pf.IDPool()
m = lambda i: vpool.id(f"m_{i}")
P = numpy.random.randint(2, size=(number_of_measurements, 2 * n), dtype=bool)
```

```python
from pysat.solvers import CryptoMinisat

with CryptoMinisat() as solver:
    for k in range(n):
        solver.add_xor_clause(
            [m(i) for i in range(number_of_measurements) if P[i, k]], value=True
        )
    for k in range(n, 2 * n):
        solver.add_xor_clause(
            [m(i) for i in range(number_of_measurements) if P[i, k]], value=False
        )
    print("Is SAT problem solvable?", solver.solve(assumptions=[m(0)]))
    print("Solution:", solver.get_model())
```
